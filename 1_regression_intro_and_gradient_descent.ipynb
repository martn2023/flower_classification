## METHOD 1A of 2: Normal Equation w/ Manual Calculation

We might not be able to leapfrog to an answer in y = mx +b format, but we can find a "mid-point" on the best-fit line by finding the average X and Y values. Then we can find calculate slope using a method described in this [YouTube vid](https://www.youtube.com/watch?v=GhrxgbQnEEU&ab_channel=EugeneO%27Loughlin). Knowing the slope (m) and a single point on the line (x and y baked in), we can isolate the b. Below is a demonstration of calculation slope "m":

<br>

y = mx +b <br>
x = x value at midpoint, so 15.60<br>
y = y value at midpoint, so 79.70<br>
m = slope of (134.8/42.4) = 3.18<br>

b = y value where x is zero, which we can calculate with the other 3 variables<br>
79.70 = (3.18 * 15.60) + b<br>
b = 79.70 - (49.61) = 30.09<br>


__Therefore, the best-fit line is y = 3.18x + 30.09__

## METHOD 1B of 2: Normal Equation, Automated w/ NumPy
I'm generating a dataset with approximately 3 as a slope and 4 as a y-intercept. The answer has zero relation to the YouTube example above because it's a different data set, so don't expect the same best-fit lines. It should be approximately 3 and 4.

Furthermore, the final best-fit line won't hit exactly, because of randomization. The final best-fit line might be something like y = 2.9(x) + 4.1

```python
'''
NumPy is a math library for moving through ARRAYS and matrix calculations faster than lists, so
the notation we see like [100,1] means 100 rows X 1 column of data points
'''
import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
X = 2 * np.random.rand(100, 1) # the 2 is the max range (from 0) on the x-axis, could have made this 10

# contrary to my intuition, we are not making 1 spreadsheet with x & y values inside. Rather, we are creating 2 separate arrays of X and Y values.
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", label="Data Points") # scatter function is looking for 2 arrays with presumably identical lengths
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Scatter Plot of Generated Data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]

theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

intercept, slope = theta_best[0, 0], theta_best[1, 0]

print(f"Intercept (b): {intercept}")
print(f"Slope (m): {slope}")
print(f"Equation of the best-fit line (random b/c it's based on random data): y = {slope:.2f}x + {intercept:.2f}")

theta_best
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color="blue", label="Data Points")

X_new = np.array([[0], [2]])
y_predict = slope * X_new + intercept
plt.plot(X_new, y_predict, "r-", label="Best-fit Line")  # Red best fit line

plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Scatter Plot with Best-fit Line")
plt.legend()
plt.grid(True)
plt.show()
```

# II. Gradient Descent
Finding the best answer through iterative experiments instead of direct calculation.

### How does it work?
<br>

"Cost" is the data science way of saying "error", or disparity between predictions and actuals. Our goal is to shift parameters around, see where the bigger drop is (we are looking for the most dramatic, steepest drop in error), and increment values towards the minimum error. The the rate which we shift parameters horizontally is known as the __"learning rate"__. A small and conservative learning rate will require more iterations to converge to the true bottom. A large and aggressive learning rate will likely step over the true minimum.

## 3x Methods of Gradient Descent
Pretend I wanted to co-found a SaaS startup, and I wanted to do some customer field studies before deploying engineering resources on building. I have 100 people who I could talk to, but I'm not sure how many product ideas will stem from these conversations, and clueless as to which idea will show the most promise.

#### 1 of 3: Batch
Reserve any conclusions until I talk to all 100. This will be slowest, but it will give the most accurate answer.

#### 2 of 3: Stochastic
Talk to one person at time, develop the product a little more after each conversation. Near the end, it might be hard to converge onto the exact best-spot.

#### 3 of 3: Mini-Batch
Split the potential customers up into 5 focus groups of 20 users each and run 5 cycles where you learn a little bit and build after. This is a blend of the 1st and 2nd method.

## Gradient Descent Calculations

#### 1 of 3: Batch
If my code works, I should get an answer identical to Normal Equation calculation

```python
X_b = np.c_[np.ones((100, 1)), X]

# learning rate: remember that smaller is more accurate but takes more iterations
eta = 0.1

n_iterations = 50

# we use m for number of data points since n reserved for iteration quantity
m = 100

# a random starting point to start exploring the cost curve
theta = np.random.randn(2, 1)
```

```python
for iteration in range(n_iterations):
    gradients = 2 / m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients # constantly recalculating theta across all seen data points

theta # don't be alarmed when it won't fit - it'll be more accurate when we increase iterations
```

#### 2 of 3: Stochastic
Batch gradient descent shown above eventually got the the correct answer once we increased the iterations enough.

However, this method is going to be a teensy bit off. Imagine you trying to 100 meters on a sandy beach from point A to point B.
- Batch would be looking at all of the footprints behind you, and drawing a best-fit line in your mind before taking your next step.
- Stochastic would be looking only at the last 2 footprints, and drawing a line based on those data points only. It'll be quicker, but dirtier.

To do a Stochastic, we're going to set __hyperparamaters__ such as:
- Epoch quantity: the number of times we do a full pass over the dataset
- An initial learning rate: the higher the number, the more aggressive we'll be with our incrementing
- Learning decay rate: the assumption is that we slow down our incrementing on a fixed schedule (not dynamically adjusted based on results). Learning rate at a certain step (t) = initial learning rate/(step number + learning decay rate).

To demonstrate how learning rate increments downwards:
- At step 0: 5/(0+50) = 0.1000
- At step 1: 5/(1+50) = 0.0980
- At step 2: 5/(2+50) = 0.0962

Now compare this if we raised the initial learning rate to 6 and/or decreased the decay to 20 (stayed aggressive longer):
- At step 0: 6/(0+20) = 0.3000
- At step 1: 6/(1+20) = 0.2857
- At step 2: 6/(2+20) = 0.2727

The first set of hyperparameters would give us a more accurate conclusion, but would take more runtime.

```python
# an epoch is a full pass over the dataset (not quite an iteration)
n_epochs = 50

# "initial learning rate"
t0 = 5 

# #learning rate decay"
t1 = 50
```

```python
def learning_schedule(t):
    return t0 / (t + t1)
 
theta = np.random.randn(2,1)  # random initialization
for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients

theta # SGD, gonna be a bit off
```

```python
# Since the end result was off by 1%, let's try more conservative hyperparameters
epoch = 500
t0 = 1 
t1 = 1000

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        
theta # more accurate than before, but still not as good as Batch Gradient Descent
```

#### 3 of 3: Mini-Batch
SGD has no hope of being 100% accurate because we're only looking at the last 2 footprints out of a trail of 100. Let's try mini-batches where I'm looking at the last 10 footprints with every next step.

Doing a batch size of only 10 yielded inaccurate results, so I pumped the batch size up to 50 (of 100 data points). The slope is fairly close, but the y-intercept is off (3.49 calculated vs 3.92 expected).

##### Warning
The below code was run many times with wildly different results, and this is outside the scope of the textbook so I'm not 100% sure my implementation was correct.

##### Takeaway
There is a noticeable accuracy benefit to Method 1 (Batch), provided we have the computing power, results aren't time sensitive, or there's a smaller data set.

```python
# Hyperparameters
n_epochs = 50
batch_size = 40 #see commentary below 
t0 = 5
t1 = 50

theta = np.random.randn(2, 1)  # Random initialization
m = 100 # amount of data points

for epoch in range(n_epochs):
    shuffled_indices = np.random.permutation(m) # Shuffling each epoch to avoid patterns
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    for i in range(0, m, batch_size):  # Go through the data in mini-batches
        xi = X_b_shuffled[i:i + batch_size]
        yi = y_shuffled[i:i + batch_size]
        
        gradients = 2 / batch_size * xi.T.dot(xi.dot(theta) - yi)
        
        eta = learning_schedule(epoch * m + i)  # learning rate temporary to this step
        
        theta = theta - eta * gradients

print(f"Theta's slope, also known m: {theta[1][0]}")
print(f"Theta's y-intercept: {theta[0][0]}")
```
