# II. Loading and Previewing Data
Using the same data set for Iris flowers we used in chapter 4

We are loading in a toy data set from the scikit-learn library, which is why the import command works

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

iris = load_iris() #essentially a dictionary used to store the data
print(iris.DESCR[:996]) # Loading the effective README that came with the data set
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
x = iris.data[:, 2:] # 4 elements, starting at index 0, but limiting this to use of only 2 features: petal length and width
y = iris.target

tree_clf_v1 = DecisionTreeClassifier(max_depth=2)
tree_clf_v1.fit(x, y) # shorthand generic naming for "tree classifier"; will not name it tree_v1 because we aren't trainimg multiple models yet
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plot_tree(
    tree_clf_v1,
    feature_names=iris.feature_names[2:],  # petal length & width
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.show()
```

```python
x = iris.data # upgrading to all 4 elements given
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x, y) # shorthand generic naming for "tree classifier"; will not name it tree_v1 because we aren't trainimg multiple models yet

plt.figure(figsize=(12, 8))
plot_tree(
    tree_clf,
    feature_names=iris.feature_names,  # petal length & width
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.show()
```
