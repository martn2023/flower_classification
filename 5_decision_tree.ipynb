# II. Loading and Previewing Data
Using the same data set for Iris flowers we used in chapter 4

We are loading in a toy data set from the scikit-learn library, which is why the import command works

We are using pandas numpy solely because of its ability to show labels. If we didn't need the labels, NumPy would be faster and more memory efficient

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

iris = load_iris() #essentially a dictionary used to store the data, this is not a pandas dataframe. Rather, its' a "Bunch" object
print(iris.DESCR[:996]) # Loading the effective README that came with the data set
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names) # using pandas, because numpy wouldn't have the labels
df['target'] = iris.target
df.head()
```


I only used pandas above in order to show the spreadsheet. The below cell never relies on the pandas dataframe I just made.


```python
x = iris.data # 4 elements, starting at index 0, but limiting this to use of only 2 features: petal length and width
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2) # what's the maximimum quantity of questions we'll allow in a data point's path before we give an answer? 2 Layers = 3 layers in the chart (2 questions and 1 outcome)
tree_clf.fit(x, y) # shorthand generic naming for "tree classifier"; will not name it tree_v1 because we aren't trainimg multiple models yet
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8)) #visualizing the tree we just made with the iris data above
plot_tree(
    tree_clf,
    feature_names=iris.feature_names,  # book only used petal length & width, but I want to use all 4 elements
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.show()
```

# III. Assessing Errors In a V1 Decision Tree
Remember that the data set was 150 rows, conveniently evened out at 50 samples for each of 3 flower species.

- Precision is "how accurate was this model when it labeled a flower type?"
- Recall is "of the total number of flowers in X species, what % were accurately detected"
- There's a natural tug-o-war going on between precision vs. recall: you can be more conservative and optimize for precision or be more aggressive and optimize for recall

### 1. Confusion Matrix With scikit-learn
 - As we did in Chapter 4 (flowers) and 3 (handwritten digits), let's get a formal *Confusion Matrix* going. It won't be native to Python.

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = tree_clf.predict(x)
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap="PuBu") #changing the look from viridis so this doesn't look like a court jester out of Disney's The Hunchback of Notre Dame
plt.show()
```

### 2. Assessing Errors In a V1 Decision Tree
- Orange box "Setosas" was flawless
- Green box Veriscolor had better recall coverage in the sense that it detected 98% (49 of 50)
- ...but purple box Virginica had better precision in the sense that 45 out of 46 claims turned out to be accurate

### 3. Automated Precision and Recall Calculations With scikit-learn
 - As we did in Chapter 4 (flowers) and 3 (handwritten digits), let's get a formal *Confusion Matrix* going. It won't be native to Python.

```python
from sklearn.metrics import classification_report
print(classification_report(y, y_pred, target_names=iris.target_names, digits=3))
```

### 4. Why Depth of 2 Instead of Something More Intense Like 5 Layers?
5 Layers would produce 100% precision and 100% recall, but that would simply be overfitting bias that wouldn't generalize to a new set of larger data points.



# IV. Entropy and Impurity Measurements
This is chaos or unpredictability in results. At every node, we should ask the question "if I divide this two nodes, will that increase accuracy?"
But impurity in itself is ***NOT*** problematic.


### 1. How Would We Quantify That?
 1) *Gini* or
 2) *Entropy* calculations

### 2. And What Are Key Differences?

Gini vs. Entropy is like mean vs. median for finding the center portion of a distributed population: not the same, but directionally accurate in most cases, at least for a novice like me. Consider the folowing:

#### A) Gini

- So by way of example, let's say:

Node #1 routes 50% of its inputs to flower A and 50% of its inputs to flower B.
1 - ((0.5)^2) - ((0.5)^2) = 1 - 0.25 - 0.25 = *0.5*

Node #2 routes 90% of its inputs to flower A and 10% of its inputs to flower B.
1 - ((0.9)^2) - ((0.1)^2) = 1 - 0.81 - 0.01 = *0.18*

Node 2 has the lower impurity score, but that doesn't mean it's doing a better job.

- This is the default method in the sci-kit learn library
